```python
endict=  dict(zip(pt['Symbol'], pt['Electronegativity']))
raddict= dict(zip(pt['Symbol'], pt['AtomicRadius']))
```

```python
elecneg=[]
atrad=[]
for i in range(len(form1)):
    electneg,atomrad=0,0
    for j in range(len(form1[i])):
#         element=form1[i][j]
        electneg+=endict[form1[i][j]]
        atomrad+=raddict[form1[i][j]]
    elecneg.append(electneg/len(form1[i]))
    atrad.append(atomrad/len(form1[i]))
        
        
```

```python
atrad
```

```python
df['wtd_en']=elecneg
df['wtd_rad']=atrad
```
